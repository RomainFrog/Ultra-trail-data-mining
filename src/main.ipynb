```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
X = pd.read_csv("../data/race.csv")
X.info()
```

```python
sns.histplot( x = X.elevation_gain)
```

```python
sns.histplot( x = X.elevation_loss)
```

```python
sns.scatterplot(
    x = X.elevation_gain,
    y = X.elevation_loss
)
```

```python
sns.scatterplot(
    x = X.elevation_gain,
    y = X.distance,
    data = X
)
```

```python
sns.scatterplot(
    x = X.distance,
    y = X.aid_stations,
    data = X
)
```

```python
Y = pd.read_csv("../data/ultra_rankings.csv")
Y.info()
```

```python
sns.histplot(x= Y.age)
```

```python
sns.boxplot(
    x="gender",
    y="age",
    data=Y
)
```

```python
sns.scatterplot(
    x = "time_in_seconds",
    y = "rank",
    data = Y,
    hue = "gender"
)
```

```python
corr = Y.corr()
sns.heatmap(corr, annot=True)
```

```python
sns.displot(
    x = "age",
    kind="hist",
    col="gender",
    data=Y,
    col_wrap=2,
    bins = 60
)
```
