```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
race = pd.read_csv("../../output/data_clean/cleaned_race.csv")
runner = pd.read_csv("../../output/data_clean/cleaned_runner.csv")
```

```python
race
```

```python
runner
```

## Univariate analysis

```python
sns.countplot(x='country', data=race)
plt.xticks(rotation=90)
plt.show()
```

```python
c = sns.countplot(x='participation', data=race)
for p in c.patches:
   c.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
sns.boxplot(y='distance', data=race)
```

```python
sns.boxplot(y='distance', data=race.loc[race.distance > 140])
```

```python
sns.boxplot(y='elevation_gain', data=race)
```

```python
sns.boxplot(y='elevation_loss', data=race)
```

```python
sns.boxplot(y='age', data=runner)
```

```python
c = sns.countplot(x='gender', data=runner)
for p in c.patches:
   c.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
c = sns.countplot(x='nationality', data=runner)
for p in c.patches:
   c.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
c = sns.countplot(x='age_range', data=runner)
for p in c.patches:
   c.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```
